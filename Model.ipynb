```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
maharashtra_data_path = '/content/Demo_census_(in).csv'
post_office_schemes_path = '/content/post_office_schemes (1).csv'

maharashtra_data = pd.read_csv(maharashtra_data_path)
post_office_schemes = pd.read_csv(post_office_schemes_path)

# Adjusting the column names of the post office schemes dataset
post_office_schemes.columns = [
    'Scheme_Name', 'Age_Min', 'Age_Max', 'Gender', 'Occupation',
    'Income_Min', 'Income_Max', 'Region'
]

# Function to recommend schemes based on correlated age, gender, and occupation demographics
def recommend_schemes(demographics, schemes):
    all_recommendations = {}
    weight_age = 0.5
    weight_gender = 0.3
    weight_occupation = 0.2
    min_score_threshold = 20.0

    for index, row in demographics.iterrows():
        district_name = row['District name']
        total_population = row['Population']
        total_workers = row['Workers']

        male_percentage = (row['Male'] / total_population) * 100
        female_percentage = (row['Female'] / total_population) * 100
        agricultural_worker_percentage = (row['Agricultural_Workers'] / total_workers) * 100

        rural_population = row['Rural_Households']
        urban_population = row['Urban_Households']

        rural_percentage = (rural_population / total_population) * 100
        urban_percentage = (urban_population / total_population) * 100

        combined_groups = {
            'Male 0-29': row['Age_Group_0_29'] * (row['Male'] / (row['Male'] + row['Female'])),
            'Female 0-29': row['Age_Group_0_29'] * (row['Female'] / (row['Male'] + row['Female'])),
            'Male 30-49': row['Age_Group_30_49'] * (row['Male'] / (row['Male'] + row['Female'])),
            'Female 30-49': row['Age_Group_30_49'] * (row['Female'] / (row['Male'] + row['Female'])),
            'Male 50+': row['Age_Group_50'] * (row['Male'] / (row['Male'] + row['Female'])),
            'Female 50+': row['Age_Group_50'] * (row['Female'] / (row['Male'] + row['Female'])),
            'Agricultural Workers': row['Agricultural_Workers']
        }

        scheme_scores = {}

        for _, scheme in schemes.iterrows():
            for group, population in combined_groups.items():
                if 'Agricultural Workers' in group and 'Kisan Vikas Patra' in scheme['Scheme_Name']:
                    if population > (total_population * 0.2):
                        if scheme['Scheme_Name'] not in scheme_scores:
                            scheme_scores[scheme['Scheme_Name']] = 0
                        scheme_scores[scheme['Scheme_Name']] += population * weight_occupation
                else:
                    gender, age_range = group.split() if 'Agricultural Workers' not in group else (None, None)

                    if age_range == '50+':
                        age_min = 50
                        age_max = float('inf')
                    elif age_range:
                        age_min = int(age_range.split('-')[0])
                        age_max = int(age_range.split('-')[1])
                    else:
                        age_min = 0
                        age_max = float('inf')

                    if scheme['Age_Min'] <= age_max and scheme['Age_Max'] >= age_min:
                        if scheme['Gender'] == 'Any' or scheme['Gender'] == gender:
                            if population > 0:
                                if scheme['Scheme_Name'] not in scheme_scores:
                                    scheme_scores[scheme['Scheme_Name']] = 0
                                score = population * (weight_age + weight_gender + weight_occupation)
                                scheme_scores[scheme['Scheme_Name']] += score

        if scheme_scores:
            max_score = max(scheme_scores.values())
            for scheme_name in list(scheme_scores):
                scheme_scores[scheme_name] = (scheme_scores[scheme_name] / max_score) * 100
                if scheme_scores[scheme_name] < min_score_threshold:
                    del scheme_scores[scheme_name]

        sorted_schemes = sorted(scheme_scores.items(), key=lambda x: x[1], reverse=True)

        all_recommendations[district_name] = {
            'combined_groups': combined_groups,
            'male_percentage': male_percentage,
            'female_percentage': female_percentage,
            'agricultural_worker_percentage': agricultural_worker_percentage,
            'rural_percentage': rural_percentage,
            'urban_percentage': urban_percentage,
            'recommendations': sorted_schemes
        }

    return all_recommendations

# Ground truth dataset
ground_truth_data = {
    'Mumbai': ['Senior Citizens Savings Scheme (SCSS)', 'Public Provident Fund (PPF)'],
    'Pune': ['National Savings Certificate (NSC)', 'Post Office Savings Account'],
    'Nagpur': ['Senior Citizens Savings Scheme (SCSS)', 'Sukanya Samriddhi Yojana (SSY)'],
    'Sindhudurg': ['Post Office Savings Account', 'Sukanya Samriddhi Yojana (SSY)']
}

def evaluate_accuracy(recommendations, ground_truth):
    total_districts = len(ground_truth)
    correct_recommendations = 0
    total_possible_recommendations = 0

    for district, actual_schemes in ground_truth.items():
        if district not in recommendations:
            print(f"Warning: '{district}' not found in recommendations. Skipping this district.")
            continue

        recommended_schemes = [rec[0] for rec in recommendations[district]['recommendations']]
        correct_recommendations += len(set(recommended_schemes) & set(actual_schemes))
        total_possible_recommendations += len(actual_schemes)

    accuracy = (correct_recommendations / total_possible_recommendations) * 100
    return accuracy

recommendations = recommend_schemes(maharashtra_data, post_office_schemes)
accuracy = evaluate_accuracy(recommendations, ground_truth_data)
print(f"Model Accuracy: {accuracy:.2f}%")

# Updated fresh light colors for pie charts
colors = ["#a8e6cf", "#dcedc1", "#ffaaa5", "#ffd3b6", "#ff8b94", "#c5cae9", "#b2dfdb"]

for district, info in recommendations.items():
    combined_dist = info['combined_groups']
    recs = info['recommendations']
    rural_percentage = info['rural_percentage']
    urban_percentage = info['urban_percentage']
    male_percentage = info['male_percentage']
    female_percentage = info['female_percentage']
    agricultural_worker_percentage = info['agricultural_worker_percentage']

    labels = list(combined_dist.keys())
    sizes = list(combined_dist.values())

    plt.figure(figsize=(10, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
    plt.title(f'Age-Gender-Agricultural Worker Distribution in {district}')
    plt.show()

    print(f"Rural Percentage in {district}: {rural_percentage:.2f}%")
    print(f"Urban Percentage in {district}: {urban_percentage:.2f}%")
    print(f"Male Percentage in {district}: {male_percentage:.2f}%")
    print(f"Female Percentage in {district}: {female_percentage:.2f}%")
    print(f"Agricultural Worker Percentage in {district}: {agricultural_worker_percentage:.2f}%")

    print(f"Recommended Schemes for {district} (Ranked by Relevance):")
    if recs:
        for scheme, relevance in recs:
            print(f" - {scheme}: {relevance:.2f}% relevance")
    else:
        print(" - No applicable schemes found.")
    print("\n")
```
